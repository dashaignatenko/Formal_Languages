```python
import nltk
from nltk import Nonterminal, nonterminals, Production, CFG
from nltk.grammar import FeatureGrammar as FCFG
import svgling
```

```python
CP, TP, vPitr, vPtr, VPtr, VPitr, NPnom, NPerg, NPobl, Vtr, Vitr, Adv, Adj, Dem, Postp, Nnom, Nobl, Nerg, Ngen = nonterminals('CP, TP, vPitr, vPtr, VPtr, VPitr, NPnom, NPerg, NPobl, Vtr, Vitr, Adv, Adj, Dem, Postp, Nnom, Nobl, Nerg, Ngen')
```

```python
grammar = CFG.fromstring("""
CP -> TP
TP -> vPitr | vPtr | nPnom vPitr | NPerg vPtr | vPitr nPnom | vPtr NPerg
vPtr -> VPtr | Adv VPtr | VPitr Adv
VPtr -> Vtr | nPnom Vtr | nPnom Vtr NPobl | Vtr nPnom | Vtr NPobl
vPitr -> VPitr | Adv VPitr | VPitr Adv
VPitr -> Vitr | Vitr NPobl | Vitr NPobl NPobl
NPobl -> Nobl Postp | Nobl
NPerg -> Nerg | Dem Nerg
nPnom -> NPnom | Dem NPnom
NPnom -> Nnom | Ngen Nnom | Nnom Ngen | Adj NPnom  | CPrel NPnom | NPnom CPrel
CPrel -> VPrel | Adv VPrel | VPrel Adv
VPrel -> Vrel | Vrel NPobl | Vrel NPobl NPobl
Nnom -> 'ŋinqej' | 'kowɬ-orw-oor' | 'ətɬa' | 'ʔəttʔə-qej' | 'kojŋə-n'
Nerg -> 'ətɬʔa-ta' | 'ʔəttʔə-qaj-a'
Nobl -> 'majŋə-ramajə-k' | 'ŋenqaj-etə' | 'ʔəttʔə-qej-e'
Ngen -> 'tʔarmas-aʔasek-en'
Vtr -> 'ne-nɬʔeten-mək' | 'rəɬʔeten-nin' | 'ɬʔu-ni-n' | 'ɣa-peɬa-ɬen'
Vitr -> 'pəkir-ɣʔ-i'
Vrel -> 'kojŋə-n-təmn-atə-ɬʔə-n'
Adv -> 'eqeɬpe' | 'n-urʔe-w'
Adj -> 'nə-mejəŋ-qin' | 'nə-ɣtin-qin' | 'simet-ɣʔ-i' | 'piŋku-sit-ɣʔ-i' | 'ɣe-wʔej-ɣərki-ɬin'
Dem -> 'ŋotqen'
Postp -> 'qasa'
""")
```

```python
def parse_chkt(sents, grammar):
    cp = nltk.parse.ChartParser(grammar)
    ts = ['ŋinqej', 'ətɬa', 'kojŋə-n', 'ʔəttʔə-qej', 'kowɬ-orw-oor', 'ətɬʔa-ta', 
          'ʔəttʔə-qaj-a', 'ʔəttʔə-qej-e', 'ŋenqaj-etə', 'majŋə-ramajə-k', 'tʔarmas-aʔasek-en', 
          'nə-mejəŋ-qin', 'nə-ɣtin-qin', 'ŋotqen', 'qasa', 'rəɬʔeten-nin', 'ɣa-peɬa-ɬen', 'ne-nɬʔeten-mək', 'kojŋə-n-təmn-atə-ɬʔə-n',
          'ɬʔu-ni-n', 'pəkir-ɣʔ-i', 'simet-ɣʔ-i', 'piŋku-sit-ɣʔ-i', 'ɣe-wʔej-ɣərki-ɬin', 'eqeɬpe', 'n-urʔe-w']
    for i in sents:
        sent = i.split()
        print(i + '\n')
        if all(w in ts for w in sent) == False:
            print('\nUnrecognized words')
        else: 
            for x in cp.parse(sent):
                print(x)
            print('\n\n')
```

```python
sents = ['ʔəttʔə-qaj-a ŋotqen nə-mejəŋ-qin kowɬ-orw-oor ɬʔu-ni-n', 'ŋinqej rəɬʔeten-nin majŋə-ramajə-k ətɬʔa-ta', 'ʔəttʔə-qaj-a ɣa-peɬa-ɬen ŋotqen eqeɬpe kojŋə-n-təmn-atə-ɬʔə-n ətɬa tʔarmas-aʔasek-en', 'ŋinqej dd-nin majŋə-ramajə-k ətɬʔa-ta']
parse_chkt(sents, grammar)
```
